```python
import plotly.express as px
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"]=(10,6)
```

```python


df=pd.read_csv("Titanic-Dataset (1).csv")
```

```python
df.head()
```

```python
df.info
```

```python
df.describe(include="all").T
```

```python
print("median values")
print(df.median(numeric_only=True))
```

```python
print("standard deviation")
print(df.std(numeric_only=True))
```

```python
numeric_cols=df.select_dtypes(include=np.number).columns
df[numeric_cols].hist(bins=20,figsize=(12,8),color="skyblue")
plt.suptitle("histogram of numeric feature")
plt.tight_layout()
plt.show()
```

```python
for col in numeric_cols:
  sns.boxplot(x=df[col])
  plt.title(f"boxplot of {col}")
  plt.show()
```

```python
sns.pairplot(df[numeric_cols].dropna())
plt.suptitle("pairplot of numeric features",y=1.02)
plt.show()
plt.figure(figsize=(10,6))
sns.heatmap(df[numeric_cols].corr(),annot=True,cmap="coolwarm",fmt=".2f")
plt.title("correlation matrix")
plt.show()
```

```python
fig=px.histogram(df,x='Sex',color="Survived",barmode="group",title="survival count by gender",labels={"survived":"Survived"})
fig.show()
```

```python
fig2=px.scatter(df,x="Age",y="Fare",color="Survived",hover_data=["Pclass"],title="age vs fare coloured by survival")
fig2.show()
```

```python
fig3=px.box(df,x="Pclass",y="Fare",color="Pclass",title="fare distribution by class")
fig3.show()
```
